#Regular expressions are really helpful when working with strings

```python
import re
```

```python
series="""
'01/18/2014 13:00:00', 100, '1st';
'01/18/2014 13:30:00', 110, '2nd';
'01/18/2014 14:00:00', 200, '3rd'
"""
```

```python
dt = re.compile("'[0-9/:\s]+'") # descibes a datetime
```

```python
result = dt.findall(series)
result
```

#this result can then be parsed and tranformed into Python datetime objects.

```python
from datetime import datetime
pydt = datetime.strptime(result[0].replace("'",""),
                        '%m/%d/%Y %H:%M:%S')
pydt
```

```python
for i in [2,3,-4,'Python', 6,-5,-4]:
    try:
        print("Square root of %d is %f" % (i, math.sqrt(i)))
    except ValueError:
        print("Cannot calculate square root of %d" % i)
    except TypeError:
        print("the input '%s' is not a numer" % i)
    else: # only if there isno exception
        print("Calculation was successful")
```

```python
for i in [2,3,-4,'Python', 6,-5,-4]:
    try:
        print("Square root of %d is %f" % (i, math.sqrt(i)))
    except ValueError:
        print("Cannot calculate square root of %d" % i)
    except TypeError:
        print("the input '%s' is not a numer" % i)
    else: # only if there isno exception
        print("Calculation was successful")
    finally: # executed in any case
        print("I am executing no matter what \n")
```
